```python
# LOADING AN IMAGE
# access image via GCollab path (folder icon)
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/CRL tutorials/photos/bauhaus_book3.jpg")
# prints image
cv2_imshow(img)
```

INTERPOLATION METHODS

```python
# SLOW INTERPOLATION
slow_res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
cv2_imshow(slow_res)
```

```python
# SHRINK INTERPOLATION
shrink_res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_AREA)
cv2_imshow(shrink_res)
```

```python
# ZOOMING INTERPOLATION
zoom_res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_LINEAR)
cv2_imshow(zoom_res)
```

TRANSLATION

```python
book2_img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/CRL tutorials/photos/bauhaus_book1.jpg")
rows,cols,ch = img.shape

M = np.float32([[1,1,10],[0,1,20]])
dst = cv2.warpAffine(book2_img,M,(cols,rows))

cv2_imshow(dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

ROTATION

```python
M = cv2.getRotationMatrix2D((cols/4,rows/2),270,1) # rotates image by 90 degree increments (90, 180, 270)
dst = cv2.warpAffine(book2_img,M,(cols,rows))

cv2_imshow(dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

AFFINE TRANSFORMATION

```python
line_img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/CRL tutorials/photos/bauhaus.jpg")
rows,cols,ch = line_img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(line_img,M,(cols,rows))

plt.subplot(121),plt.imshow(line_img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

PERSPECTIVE TRANSFORMATION

```python
book2_img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/CRL tutorials/photos/bauhaus_book2.jpg")
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(book2_img,M,(300,300))

plt.subplot(121),plt.imshow(book2_img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
